```python
# from google.colab import files
# uploaded = files.upload()
import pandas as pd


mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

print("Data Frame")
myvar = pd.DataFrame(mydataset)
print(myvar)

print("Series") #series means 1d array holding data of any type
mydataset = [i for i in range(1,6)]
myvar = pd.Series(mydataset)
print(myvar)
print(myvar[3]) #label

#create labels
myvar = pd.Series(mydataset, index = ["a", "b", "c", "d", "e"])
print(myvar)
print(myvar['a'])

#key value object series
mydataset = {
  'name': "duck",
  'age': 22,
  'city': "Bangalore"
}

#key of dictionary becomes label
myvar = pd.Series(mydataset)
print(myvar)

#select specific index from dictionary
d = pd.Series(mydataset, index = ["name", "city"])
print(d)
```

```python
#cleaning - fixing bad data in my data set

#bad data could be,
'''
Empty cells
Data in wrong format
Wrong data
Duplicates

Note:
By default, the dropna() method returns a new DataFrame, and will not change the original.
If you want to change the original DataFrame, use the inplace = True
'''
# from google.colab import files
# uploaded = files.upload()
data = uploaded['mydata (4).csv']
df = pd.read_csv(io.StringIO(data.decode('utf-8')))
print(df.dropna()) #remove rows with empty cells
print('-'*30)

# print(df.dropna(inplace = True)) #remove rows with empty cells and change original dataframe
'''
Note: Now, the dropna(inplace = True) will NOT return a new DataFrame,
but it will remove all rows containing NULL values from the original DataFrame.
'''
# print(df)
#replace empty cells with 0
# df.fillna(0, inplace = True)

#replace only for specific columns
df.fillna({"Age" : 0}, inplace = True)
print(df)

#mean, median, mode

# Mean = the average value (the sum of all values divided by number of values)
# Median = the value in the middle, after you have sorted all values ascending.
# Mode = the value that appears most frequently.

x = df["Age"].mean()
x = df["Age"].median()
x = df["Age"].mode()[0] #mode returns a series so we need to access the first element
# df["Age"].fillna(x, inplace = True)
print(x)


```
